```python
import numpy as np
import matplotlib.pyplot as plt
```

$$ \vec{f}(x,y) = \sin(x)\, \hat{x} + \cos(y)\, \hat{y} $$

```python
from numpy.lib.function_base import meshgrid
x_range = y_range = np.linspace(0,2*np.pi,10)
x, y = np.meshgrid(x_range, y_range)

fx, fy = np.sin(x), np.cos(y)
```

```python
plt.figure(figsize=(6,6))
Q = plt.quiver(x, y, fx,fy, pivot="middle")
plt.quiverkey(Q, .85, 1.01, 1,"= 1 m/s", labelpos="E")

plt.title("Flow")
plt.axis("scaled")
plt.show()
```

```python
plt.streamplot(x,y,fx, fy, density=2)
plt.axis("scaled")
plt.show()
```

```python
x_range = y_range = np.linspace(0,2*np.pi,100)
x, y = np.meshgrid(x_range, y_range)
fx, fy = np.sin(x), np.cos(y)
f_mag = np.sqrt(fx**2+ fy**2)

plt.figure(figsize=(8,8))
plt.imshow(f_mag, origin="lower", extent=(0,2*np.pi,0,2*np.pi), cmap="gray")
plt.quiver(x[5::10,5::10], y[5::10,5::10], fx[5::10,5::10], fy[5::10,5::10], pivot = "middle", color = "red")
plt.show()
```

$$ \vec{\nabla}f(x,y) = \frac{\partial f}{dx}\, \hat{x} + \frac{\partial f}{dy}\, \hat{y}$$

```python
def f(x,y):
  return np.cos(x)**2 * np.cos(y)**2 + y
```

```python
x_range = y_range = np.linspace(-np.pi/2, np.pi/2, 101)
x, y = np.meshgrid(x_range,y_range)

dfdy, dfdx = np.gradient(f(x,y), np.pi/100)
```

```python
plt.figure(figsize=(8,8))
plt.contour(f(x,y),  extent=(-np.pi/2,np.pi/2,-np.pi/2,np.pi/2))
plt.quiver(x[5::10,5::10],y[5::10,5::10],dfdx[5::10,5::10],dfdy[5::10,5::10], pivot="middle")
plt.show()
```
